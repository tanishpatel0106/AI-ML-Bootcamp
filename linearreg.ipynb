Step 1: Generating the Dataset

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data
X = 2 * np.random.rand(100, 1)  # 100 samples from 0 to 2
true_slope = 3.5
true_intercept = 1.25
y = true_slope * X + true_intercept + np.random.randn(100, 1)  # y = 3.5x + 1.25 + noise

# Visualize
plt.scatter(X, y, color='blue')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Linear Dataset")
plt.show()
```

Step 2: Using Scikit Learn Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Create model and fit
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Coefficients
print("Slope (Weight):", model.coef_[0][0])
print("Intercept (Bias):", model.intercept_[0])

# Plot
plt.scatter(X, y, label='Data', alpha=0.5)
plt.plot(X, y_pred, color='red', label='Linear Fit')
plt.legend()
plt.title("Scikit-learn Linear Regression")
plt.show()
```

Step 3: Using Manual Gradient Descent

Step 3.1 - Making the Loss Function

```python
# Add a column of ones to X for bias term
X_b = np.c_[np.ones((100, 1)), X]  # shape (100, 2)

# Initialize weights randomly (theta_0 = intercept, theta_1 = slope)
theta = np.random.randn(2, 1)  # shape (2, 1)

# Learning rate and iterations
learning_rate = 0.1
n_iterations = 1000
m = len(X_b)

# Gradient Descent Loop
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients

print("Manually Trained Intercept:", theta[0][0])
print("Manually Trained Slope:", theta[1][0])
```

Step 3.2 - Plotting the GD

```python
# Predictions
y_pred_manual = X_b.dot(theta)

# Plot
plt.scatter(X, y, alpha=0.5, label='Data')
plt.plot(X, y_pred_manual, color='green', label='Gradient Descent Fit')
plt.legend()
plt.title("Manual Gradient Descent Linear Regression")
plt.show()
```
